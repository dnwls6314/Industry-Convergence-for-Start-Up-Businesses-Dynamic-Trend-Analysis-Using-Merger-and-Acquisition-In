```python
import pandas as pd
```

```python
doc=pd.read_csv("./2020년데이터/2020.csv",header=0, encoding='utf8',error_bad_lines=False)
doc
```

```python
acquiree_industry=list(doc['Acquiree Industries'])
```

```python
acquiree_industry
```

```python
acquiree_industry_=[]
for a in acquiree_industry:
    b=str(a).split(', ')
    acquiree_industry_.append(b)
```

```python
acquiree_industry_
```

```python
import itertools
acquiree_industry__=list(itertools.chain.from_iterable(acquiree_industry_))
```

```python
acquiree_industry__=pd.DataFrame(acquiree_industry__, columns=['acquiree industry'])
acquiree_industry__
```

```python
acquiree_industry__.to_csv("./2020년_topN_피인수산업_통계.csv", encoding='utf8')
```
